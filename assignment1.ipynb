### Relu activation curve

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.linspace( -100,100, 201)
Y = np.maximum(0,X)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(X, Y, linewidth=2, label="ReLU(x)")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)

plt.title("ReLU Activation Function (-100 to 100)", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("Y", fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()
```

### Tanh and sigmoid function curves

```python
xs = np.arange(-100,100,0.25)

def tanh(x):
    return (np.exp(2*x) - 1)/(np.exp(2*x) + 1)

def sigmoid(x):
    return 1/(1 + np.exp(-x))

ys1 = tanh(xs)
ys2 = sigmoid(xs)


plt.plot(xs,ys1,label = 'Tanh Activation curve', s = 6, c = 'blue')
plt.plot(xs,ys2,label="Sigmoid Curve", c= "red", s = 6)
```
